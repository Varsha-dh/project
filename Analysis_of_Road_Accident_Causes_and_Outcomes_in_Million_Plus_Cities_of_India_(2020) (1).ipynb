```python
plt.figure(figsize=(10, 6))  # Set figure size
sns.countplot(data=df, x='Outcome of Incident',hue='Cause category')  # Create countplot with hue
plt.title('Cause Category by Outcome of Incident ')  # Set title
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()  # Display the plot
```

"The most common cause category of incidents is related to road features and the impacting vehicle."

```python
plt.figure(figsize=(10, 6))  # Set figure size
sns.countplot(data=df, x='Cause Subcategory')  # Create countplot with hue
plt.title('Cause Subcategory')  # Set title
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()  # Display the plot
```
